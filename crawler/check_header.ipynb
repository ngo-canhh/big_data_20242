```python
import pandas as pd
import time
def compare_csv_columns(file1_path, file2_path):
    try:
        # Đọc hai file CSV
        df1 = pd.read_csv(file1_path, nrows=0)  # Chỉ đọc header
        df2 = pd.read_csv(file2_path, nrows=0)  # Chỉ đọc header

        # Lấy danh sách cột
        columns1 = set(df1.columns)
        columns2 = set(df2.columns)

        # So sánh cột
        if columns1 == columns2:
            print("Hai file CSV có cùng các cột.")
        else:
            print("Hai file CSV có các cột khác nhau.")
            # Cột chỉ có trong file 1
            only_in_file1 = columns1 - columns2
            if only_in_file1:
                print(f"Cột chỉ có trong file 1: {only_in_file1}")
            # Cột chỉ có trong file 2
            only_in_file2 = columns2 - columns1
            if only_in_file2:
                print(f"Cột chỉ có trong file 2: {only_in_file2}")

    except FileNotFoundError as e:
        print(f"Lỗi: Không tìm thấy file - {e}")
    except Exception as e:
        print(f"Lỗi: {e}")

# Ví dụ sử dụng
file1_path = "Final_Data.csv"
file2_path = "Cleaned_Data.csv"
compare_csv_columns(file1_path, file2_path)
#Thêm vào file2 các các cột bị thiếu trong file 1 và gán giá trị mặc định là 0
def add_missing_columns(file1_path, file2_path):
    try:
        # Đọc hai file CSV
        df1 = pd.read_csv(file1_path)
        df2 = pd.read_csv(file2_path)

        # Lấy danh sách cột
        columns1 = set(df1.columns)
        columns2 = set(df2.columns)

        # Tìm các cột chỉ có trong file 1
        missing_columns = columns1 - columns2

        # Thêm các cột thiếu vào df2 với giá trị mặc định là 0
        for column in missing_columns:
            if column != 'crawl_timestamp':
                df2[column] = 0
            else:
                df2[column] = int(time.time()* 1000) 
        # Lưu lại file CSV mới
        df2.to_csv(file2_path, index=False)
        print(f"Đã thêm các cột thiếu vào {file2_path} và gán giá trị mặc định là 0.")

    except FileNotFoundError as e:
        print(f"Lỗi: Không tìm thấy file - {e}")
    except Exception as e:
        print(f"Lỗi: {e}")

add_missing_columns(file1_path, file2_path)

# Xóa cột source ở trong cả hai file
def remove_source_column(file1_path, file2_path):
    try:
        # Đọc hai file CSV
        df1 = pd.read_csv(file1_path)
        df2 = pd.read_csv(file2_path)

        # Xóa cột 'source' nếu nó tồn tại
        if 'source' in df1.columns:
            df1.drop(columns=['source'], inplace=True)
        if 'source' in df2.columns:
            df2.drop(columns=['source'], inplace=True)

        # Lưu lại file CSV mới
        df1.to_csv(file1_path, index=False)
        df2.to_csv(file2_path, index=False)
        print("Đã xóa cột 'source' trong cả hai file.")

    except FileNotFoundError as e:
        print(f"Lỗi: Không tìm thấy file - {e}")
    except Exception as e:
        print(f"Lỗi: {e}")
remove_source_column(file1_path, file2_path)
```

```python
#Xoá các cột có player-agent bị null trong file 2
def remove_null_user_agent(file2_path):
    try:
        # Đọc file CSV
        df2 = pd.read_csv(file2_path)

        # Xoá các cột có giá trị 'player-agent' là null
        if 'player-agent' in df2.columns:
            df2 = df2[df2['player-agent'].notnull()]

        # Lưu lại file CSV mới
        df2.to_csv(file2_path, index=False)
        print("Đã xoá các hàng có 'player-agent' là null trong file.")

    except FileNotFoundError as e:
        print(f"Lỗi: Không tìm thấy file - {e}")
    except Exception as e:
        print(f"Lỗi: {e}")
```

```python
import pandas as pd
import logging

# --- Cấu hình Logging ---
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[logging.StreamHandler()]
)
logger = logging.getLogger(__name__)

# --- Hàm xóa cột trong file CSV ---
def remove_column_from_csv(input_csv, output_csv, column_name):
    try:
        # Đọc file CSV
        df = pd.read_csv(input_csv, encoding='utf-8-sig')
        logger.info(f"Loaded {len(df)} records from {input_csv}")
        
        # Kiểm tra xem cột có tồn tại không
        if column_name not in df.columns:
            logger.error(f"Column '{column_name}' not found in {input_csv}")
            return False
        
        # Xóa cột
        df = df.drop(columns=[column_name])
        logger.info(f"Removed column '{column_name}'")
        
        # Lưu vào file CSV mới
        df.to_csv(output_csv, index=False, encoding='utf-8-sig')
        logger.info(f"Saved modified data to {output_csv}")
        return True
    
    except FileNotFoundError:
        logger.error(f"File {input_csv} not found")
        return False
    except Exception as e:
        logger.error(f"Error processing file {input_csv}: {e}")
        return False

# --- Hàm chính ---
def main():
    # Đường dẫn file CSV
    input_csv = "Cleaned_Data.csv"  # Thay bằng đường dẫn tới file CSV của bạn
    output_csv = "Cleaned_Data.csv"     # File CSV đầu ra
    column_to_remove = "player_agent"          # Tên cột muốn xóa (ví dụ: player_agent)

    # Gọi hàm xóa cột
    if remove_column_from_csv(input_csv, output_csv, column_to_remove):
        logger.info("Column removed successfully")
    else:
        logger.error("Failed to remove column")

if __name__ == "__main__":
    main()
```
